## after sorting

```python
template = np.full((720, 1080, 3), (0,255,0), dtype='uint8')
```

```python
pts = [[272, 545],
       [481, 341],
       [485, 458],
       [274, 418]]
```

```python
len(pts)
```

```python
# do this if length of list is 4
# sort the complete list by x in increasing order
pts = sorted(pts, key=lambda x: x[0])
# sort first 2 elements of the list by y in decreasing order
pts_1 = sorted(pts[:2], key=lambda x: x[1], reverse=True) 
# sort last 2 elements of the list by y in increasing order
pts_2 = sorted(pts[2:], key=lambda x: x[1])
# combine the two lists 
new_list = pts_1+pts_2
new_list
```

```python
new_list[:2]
```

```python
pts_1
```

```python
new_list = np.array(new_list)

new_list = new_list.reshape((-1, 1, 2))

image = cv2.polylines(template, [new_list],
                      isClosed=True, color=(255,0,0), thickness=4)
```

```python
plt.imshow(image)
```

```python
image = cv2.line(image, pts_1[0], pts_2[1], (0,0,0), 3)
plt.imshow(image)
```

```python
pts_1[0]
```

```python
pts_2[1]
```

```python
p1 = pts_1[0]
p2 = pts_2[1]
```

```python
theta = np.arctan2(p1[1]-p2[1], p1[0]-p2[0])
endpt_x = int(p1[0] - 1000*np.cos(theta))
endpt_y = int(p1[1] - 1000*np.sin(theta))
```

```python
cv2.line(image, (p1[0], p1[1]), (endpt_x, endpt_y), 255, 2)
plt.imshow(image)
```

```python
start_x = int(p1[0] + 1000*np.cos(theta))
start_y = int(p1[1] + 1000*np.sin(theta))

cv2.line(image, (start_x, start_y), (endpt_x, endpt_y), 255, 2)
plt.imshow(image)
```
